```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcdefaults()
```

```python
plt.style.use("bmh")
```

```python
price = [15, 25]
```

```python
sold = [50, 200]
```

```python
plt.scatter(price[0], sold[0])
plt.scatter(price[1], sold[1])
plt.xlim(0, 35)
plt.ylim(0, 250)
plt.annotate("  Before\nPandemic", (10, 30), color='b')
plt.annotate("  After\nPandemic", (21, 180), color='r')
plt.xlabel("Price")
plt.ylabel("Quantity")
#plt.tight_layout()
plt.savefig("supply and demand.png", dpi=900)
plt.show()
```

```python
price2 = [7,9,11,13,15,17,19]
```

```python
x = "41 37 33 29 25 21 17".split()
```

```python
quant = [int(a) for a in x]
```

```python
plt.plot(price2, quant, color='c', label='Demand')
plt.xlim(5, 20)
plt.ylim(15)
plt.xlabel("Price")
plt.ylabel("Quantity")
plt.axhline(25, xmin=.47, xmax=.665, linestyle='--')
plt.axvline(12, ymin=.369, ymax=.585, linestyle='--')
plt.text(13, 23.5, "ΔX")
plt.text(11.25, 27.5, "ΔY")
plt.text(7, 22, r"$Slope\ =\ \frac{\ \Delta y\ }{\ \Delta x\ }$", size=15, color='b')
plt.text(13, 35, r"$E_d\ =\ \frac{\ \%\Delta Q\ }{\ \%\Delta P\ }$", size=15, color='r')
plt.legend()
plt.tight_layout()
#plt.savefig("demand curve.png", dpi=900)
plt.show()
```
